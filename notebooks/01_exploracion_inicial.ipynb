**Objetivo de este notebook:** Carga inicial de los 10 CSV, visualización de primeras filas, tipos de datos y métricas básicas de calidad.


```python
import pandas as pd
import os

# ─── Ajusta esta ruta si es necesario ───────────────────────────────────────
DATA_DIR = "../data"
# ─────────────────────────────────────────────────────────────────────────────

CSV_FILES = {
    "matches":          "charting-m-matches.csv",
    "overview":         "charting-m-stats-Overview.csv",
    "serve_basics":     "charting-m-stats-ServeBasics.csv",
    "serve_direction":  "charting-m-stats-ServeDirection.csv",
    "return_depth":     "charting-m-stats-ReturnDepth.csv",
    "return_outcomes":  "charting-m-stats-ReturnOutcomes.csv",
    "rally":            "charting-m-stats-Rally.csv",
    "net_points":       "charting-m-stats-NetPoints.csv",
    "kp_serve":         "charting-m-stats-KeyPointsServe.csv",
    "kp_return":        "charting-m-stats-KeyPointsReturn.csv",
}

print(f"Directorio de datos: {os.path.abspath(DATA_DIR)}")
print(f"Archivos esperados: {len(CSV_FILES)}")
```

```python
# Carga de todos los CSV con manejo de errores de encoding
dfs = {}

for key, filename in CSV_FILES.items():
    filepath = os.path.join(DATA_DIR, filename)
    try:
        df = pd.read_csv(filepath, encoding="utf-8", low_memory=False)
    except UnicodeDecodeError:
        df = pd.read_csv(filepath, encoding="latin-1", low_memory=False)
    dfs[key] = df
    print(f"✓ {filename:<50} → {df.shape[0]:>6} filas × {df.shape[1]:>3} columnas")
```

## Resumen de calidad — todas las tablas

```python
resumen = []

for key, df in dfs.items():
    nulos_total = df.isnull().sum().sum()
    nulos_pct   = round(nulos_total / (df.shape[0] * df.shape[1]) * 100, 2)
    duplicados  = df.duplicated().sum()
    resumen.append({
        "Tabla":        key,
        "Filas":        df.shape[0],
        "Columnas":     df.shape[1],
        "Nulos totales": nulos_total,
        "% Nulos":      nulos_pct,
        "Duplicados":   duplicados,
    })

pd.DataFrame(resumen).set_index("Tabla")
```
